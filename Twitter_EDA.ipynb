## Adidas

```python
df_adidas = pd.read_pickle("./Pickles/df_official_adidas_tweets.pkl")
```

```python
df_adidas
```

```python
df_adidas.info()
```

```python
hashtags = []
for hashtag in df_adidas['hashtags'].values.tolist():
    if len(hashtag) != 0:
        hashtags.extend(hashtag)
```

```python
hashtags_dict = {x: hashtags.count(x) for x in hashtags}
```

```python
adidas_tag_count = pd.DataFrame.from_dict(hashtags_dict, orient='index', columns=['tag_count'])
```

```python
adidas_tag_count = adidas_tag_count.sort_values(by=['tag_count'], ascending=False)
```

```python
adidas_tag_count
```
