```python
# Imports and installs

import pandas as pd
pd.set_option('max_columns', 30)
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style='whitegrid', palette='bright', color_codes=True)
sns.set_context('talk')
import pickle
```

## Adidas

```python
with open('./Pickles/all_adidas_scores.pickle', 'rb') as f:
    all_adidas_scores = pickle.load(f)
```

```python
len(all_adidas_scores)
```

```python
df_all_adidas_scores = pd.DataFrame(all_adidas_scores)
```

```python
df_all_adidas_scores
```

```python
df_all_adidas_scores.drop_duplicates(subset='name', inplace=True)
```

```python
df_all_adidas_scores
```

```python
with open('./Pickles/all_adidas_summaries.pickle', 'rb') as f:
    all_adidas_summaries = pickle.load(f)
```

```python
df_all_adidas_summaries = pd.DataFrame(all_adidas_summaries)
```

```python
len(all_adidas_summaries)
```

```python
df_all_adidas_summaries.drop_duplicates(subset='name', inplace=True)
```

```python
df_all_adidas_summaries
```

```python
df_adidas = pd.merge(df_all_adidas_scores ,df_all_adidas_summaries, on='name')
```

```python
df_adidas
```

```python
# Changed score to int64

df_adidas['score'] = df_adidas['score'].astype('int64')
```

```python
df_adidas.dtypes
```

```python
df_adidas.describe()
```

```python
df_adidas.groupby('score').count()[['name']]
```

```python
df_adidas.plot()
```

```python
df_adidas.hist()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df_adidas.corr(), annot=True)
```
