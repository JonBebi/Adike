```python
with open('./Pickles/sneakers_prices_users.pickle', 'rb') as f:
    sneakers_prices_users = pickle.load(f)
```

```python
df_sneakers_prices_users = pd.DataFrame(sneakers_prices_users)
```

```python
df_sneakers_prices_users.head(10)
```

```python
df_sneakers_prices_users.tail(10)
```

```python
df_sneakers_prices_users[df_sneakers_prices_users['price'] == 'price'].count()
```

```python
# Dropped sneakers with no price

df_sneakers_prices_users = df_sneakers_prices_users[df_sneakers_prices_users['price'] != 'price']
```

```python
df_sneakers_prices_users[df_sneakers_prices_users['price'] == 'price'].count()
```

```python
# Removed '$'

df_sneakers_prices_users['price'] = df_sneakers_prices_users['price'].str[1:]
```

```python
# Changed price to int64

df_sneakers_prices_users['price'] = df_sneakers_prices_users['price'].astype('int64')
```

```python
# Removed whitespace

df_sneakers_prices_users['users'] = df_sneakers_prices_users['users'].str.strip()
```

```python
# Removed ' users'

df_sneakers_prices_users['users'] = df_sneakers_prices_users['users'].str[:-6]
```

```python
# Changed users to int64

df_sneakers_prices_users['users'] = df_sneakers_prices_users['users'].astype('int64')
```

```python
df_sneakers_prices_users
```

```python
# Added brand column

df_sneakers_prices_users['brand'] = ['Adidas' if i <= 382 else 'Nike' for i in df_sneakers_prices_users.index]
```

```python
df_sneakers_prices_users
```

```python
df_sneakers_prices_users.groupby('brand').count()
```

```python
df_sneakers_prices_users.groupby(['price', 'brand']).count()
```

```python
# Adidas df with aggregated data

adidas_price = df_sneakers_prices_users.groupby(['brand', 'price']).count().iloc[:57]
```

```python
# Nike df with aggregated data

nike_price = df_sneakers_prices_users.groupby(['brand', 'price']).count().iloc[57:]
```

```python
adidas_price
```

```python
# Reseted index

adidas_price1 = adidas_price.reset_index()
```

```python
adidas_price1
```

```python
adidas_price1.columns
```

```python
# Dropped unnecessary columns

del adidas_price1['brand']
del adidas_price1['users']
```

```python
adidas_price1
```

```python
# Reseted index

nike_price1 = nike_price.reset_index()
```

```python
nike_price1
```

```python
nike_price1.columns
```

```python
# Dropped unnecessary columns

del nike_price1['brand']
del nike_price1['users']
```

```python
nike_price1
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.barplot(x=adidas_price1['price'], y=adidas_price1['name'], color='black', label='Adidas')
sns.barplot(x=nike_price1['price'], y=nike_price1['name'], color='grey', label='Nike')
ax.set_xlabel("Price of sneaker")
ax.set_ylabel("# of sneakers with specific price")
plt.xticks(rotation=90)
ax.legend()
```
